```python
#7 print the model whose horsepower is the maximum

df['Model'].agg({'Horsepower':'max'})
```

```python
#8 drop column with the highest number of NA. DON'T make the change permanent 


m=df.isna().sum().max()
condition=(df.isna().sum()==m)
df.drop((df.columns[condition])[0], axis=1)



```

```python
#9 show the name of manufacture whose sales were highest

df['Manufacturer'].agg({'Sales_in_thousands':'max'})
```

```python
#10 show data of cars whose latest launch was in 2011 
df['Latest_Launch'] = pd.to_datetime(df['Latest_Launch'])
df['year']=df['Latest_Launch'].dt.year
condition= df['year']==2011
df[condition]
```

```python
#11 show data of cars whose fuel efficiency was between 10 to 15
df[df['Fuel_efficiency'].between(10,15)]
```
